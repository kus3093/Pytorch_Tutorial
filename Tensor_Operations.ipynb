```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# reshape and view
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# resape if we don't know the number of items using -1
my_torch2 = torch.arange(15)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(-1,3)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch3
```

```python
my_torch4 = my_torch3.view(2,5)
my_torch4
```

```python
# with reshape and view they will update
my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 4141
my_torch5
```

```python
my_torch6
```

```python
#Slices
my_torch7 = torch.arange(10)
my_torch7
```

```python
my_torch7[7]
```

```python
# Grab a Slice
my_torch8 = my_torch7.reshape(5,2)
my_torch8
```

```python
my_torch8[:,1]
```

```python
# return column
my_torch8[:,1:]
```
